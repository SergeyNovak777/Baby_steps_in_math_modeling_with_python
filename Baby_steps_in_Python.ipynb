## ***Условные операторы, циклы, Comprehensions***

### Условные операторы

```python
# условные операторы имеют следующий вид

#   if условие
#       код, который выполняется при выполнении условия
#   else
#        код, который выполняется при не выполнении условия
```

```python
# расширенная конструкция условного оператора

#   if условие 1
#       код, который выполняется при выполнении условия 1
#   elif условие 2
#        код, который выполняется при выполнении условия 2
#   else
#       код, который выполняется при не выполнении условия 2
```

```python
# Пример использования условного оператора

def tellNumber(N):
    if N > 5:
        print("It is {} and it is >5".format(N));
    elif N <4:
        print("It is {} and it is <4".format(N));
    elif N == 5:
        print("It is five.");
    else:
        print("FOUR!");

tellNumber(2)
```

### Циклы

Примеры циклов в Python

```python
someList = [1, 2., [3, 4.]];

for val in someList:
    print(val);
# val является элементов в someList
```

```python
# То же самое, что и выше, но с синтаксиксом похожим на C/C++
for i in range(0, len(someList),  1):
    print(someList[i]);
```

```python
for i, val in enumerate(someList):
    print("At {} we have {}".format(i, val));
# i это индекс элемента val
```

**Функция zip**  берёт на вход несколько списков и создаёт из них список (в Python 3 создаётся не list, а специальный zip-объект) кортежей, такой, что первый элемент полученного списка содержит кортеж из первых элементов всех списков-аргументов.

```python
animals = ['Dog', 'Cat', 'Bird', 'Mouse'];
actions = ['"woof"', '"meow"', '"tweet"', '"squeak"'];

for animal, action in zip(animals, actions):
    print("{} goes {}".format(animal, action));
```

```python
for val in zip(animals, actions):
    animal, action = val;
    print("{} goes {}".format(animal, action));
```

### Comprehensions

В синтаксисе Python есть удобные конструкции, которые помогают быстро преобразовать один список в другой. В некотором смысле, эта конструкция в большинстве ситуаций обобщает циклы и является более компактной и понятной версией их.

**Рассмотрим comrehensions на примере задачи:** есть список чисел от 1 до 100, из которого нужно выбрать все нечётные числа от 10 до 98 и возвести их в квадрат

```python
# Решение задачи с помощью циклов

numberList = range(1, 101);

resList = [];
for n in numberList:
    if (10<= n and n<= 98) and (n % 2 ==1):
        resList.append(n**2);
```

```python
# Решение задачи с помощью comrehensions
resList2 = [n**2 for n in numberList if (10<= n and n<= 98) and (n % 2 ==1)];
```

## ***Простые графики***

```python
# импортирование пакетов
import numpy as np
import matplotlib.pyplot as plt
# import является ключевым словом после которого идет наименование пакета
# ключевое слово as является необязательным и позволяет как бы переименовать пакет
# именем, которое следует после as и обращаться к функциям в нем по этому слову
# matplotlib.pyplot позволяет импортировать пакет не целиком, а импортирует только модуль pyplot
```

### Построение графиков

```python
# np.linspace(start, stop, num) создает массив в пределах [start; stop] и разбивает его на num значений
xs = np.linspace(-2, 2, 1001)
# здесь создается массив от -2 до 2 и разбивается на 1001 значение

ys = [x*(1-x)*(1+x) for x in xs]
# здесь используется Comprehensions, который подставляет каждое значение из xs в формулу x*(1-x)*(1+x)
# и в результате получается некоторый массив
```

```python
# построение графика
# plot соединяет точки между собой кривой
plt.plot(xs, ys)
```

```python
# plot может принимать только один массив, в таком случае ось х будет соответствовать индексам, а ось y в данном случае ys
plt.plot(ys)
```

### Отрисовка точек

Для рисования точек используется функция **scatter**. Она аналогична функции plot и имеет аналогичные аргументы. Для задания размера точки используется
**s**, для задания цвета используется **с**.
Подробнее о [np.random.normal](https://numpy.org/doc/stable/reference/random/generated/numpy.random.normal.html)

```python
xvals = np.linspace(-1, 1., 1001)
y1vals = np.array([2*x*(1-x**2) for x in xs])
# np.array задает массив и заполняет его согласно [2*x*(1-x**2) for x in xs]

y2vals = y1vals + np.random.normal(size=len(y1vals))
# np.random.norma случайные выборки из нормального Гауссовского распределения
plt.plot(xvals, y1vals, 'k--', label='Graph')
skip = 5
plt.scatter(xvals[::skip], y2vals[::skip], s=4,c='#da9032', label='Random fluctuations')
plt.legend()
```

### Отрисовка функций двух переменных

Для отрисовки функций двух переменных используется функция **pcolormesh**. Она принимает три массива данных x, y, z

```python
xs = ys = np.linspace(0.0,1.0, 801)
grid = np.zeros([len(ys), len(xs)])
# np.zeros задает нулевой массив размерами len(ys) x len(xs)
for i, y in enumerate(ys):
    for j, x in enumerate(xs):
        grid[i][j] = np.log(1e-4+(x-0.2)**2+(y-0.6)**2/4)
```

```python
plt.pcolormesh(xs,ys,grid,cmap=plt.cm.get_cmap('RdBu'))
# cmap задает цветовую карту
plt.colorbar()
# plt.colorbar() отрисовывает цветовую шкалу
```

## ***Поиск состояний равновесия***

Информацию о состояниях равновесия можно найти в [задачнике Филлипова на страницу 98 в разделе особые точки](https://kvm.gubkin.ru/pub/uok/FilippovDU.pdf)

Рассмотрим систему:
$$ \dot{\rho}_1 = \rho_1 (\delta_1 - d \rho_2^2 - \rho_1^2),\;\; \dot{\rho}_2 = \rho_2 (\delta_2 - d \rho_1^2 - \rho_2^2)$$

```python
# is_negative = False означает, что переменные не могут быть отрицательными,
# is_real =True переменные могут быть только вещественными

r1, r2 = sp.symbols('rho1 rho2', is_negative=False, is_real =True)
delta1, delta2 = sp.symbols('delta_1 delta_2', is_real =True)
d = sp.symbols('d', is_real =True)

rhs1 = r1*(delta1 - d*r2**2 - r1**2)
rhs2 = r2*(delta2 - d*r1**2 - r2**2)
```

```python
# Для поиска состояний равновесий мы используем команду solve
sols1 = sp.solve([rhs1, rhs2], [r1, r2], dict=True)
# dict=True означает, что ответ будет хранится в виде словаря
# о том, что такое словарь читайте в разделе Основы Python
```

```python
# исключаем тривиальные решения
sols2 = sp.solve([rhs1/r1, rhs2/r2], [r1, r2], dict=True)
sols2
```

***Проверим последнеее состояние равновесия, подставив его в исходную систему***

Запишем координаты состояния равновесия в отдельную переменную

```python
eq1 = sols2[-1]; eq1
```

```python
eq1r1 = eq1[r1]; eq1r1
```

```python
eq1r2 = eq1[r2]; eq1r2
```

Составим матрицу из правых частей системы

```python
rhs = sp.Matrix([rhs1, rhs2]); rhs
```

Подставим в правую часть место p1, p2 координаты состояния равновесия и упростим выражение

```python
# с помощью subs выполняется подстановка
(rhs).subs({r1: eq1r1, r2: eq1r2})
```

```python
sp.expand(rhs).subs({r1: eq1r1, r2: eq1r2})
```

```python
sp.expand(sp.expand(rhs).subs({r1: eq1r1, r2: eq1r2}))
```

```python
sp.simplify(sp.expand(sp.expand(rhs).subs({r1: eq1r1, r2: eq1r2})))
```

Как видно выше получили нули, следовательно состояние равновесия найдено верно

## ***Матрица Якоби***

```python
# составим Якобиан
Jac = rhs.jacobian([r1, r2]); Jac
```

```python
# Подставляем в Якобиан состояние равновесия
Jac.subs(eq1)
```

```python
# Упрощаем полученную матрицу
JacAtSaddle = sp.simplify(sp.expand(Jac.subs(eq1))); JacAtSaddle
```

```python
# Попробуем найти собственные числа, иногда это получается. Иногда потому, что системы как правило не поддаются аналитическому исследованию
JacAtSaddle.eigenvals()
```

# ***Численное решение с помощью scipy***

В большинестве случаев аналитический анализ системы ОДУ невозможен. В этом случае применяются численные методы. В данной части рассмотрим как получить численное решение системы ОДУ с помощью пакета scipy. С численными методами интегрирования можно ознакомиться [в документации пакета](https://docs.scipy.org/doc/scipy/reference/generated/scipy.integrate.solve_ivp.html)
Особенно в них можно не вдаваться, информации с раздела будет достаточно.

Рассмотрим уравнение Ван дер Поля:
$$ \ddot{x} - \varepsilon \dot{x} (1 - x^2) + x  = 0$$

```python
def vanDerPol(t, X):
    x, v = X
    return [v, -x + 0.15*v*(1-x**2)]
```

```python
from scipy.integrate import solve_ivp
```

```python
# Время в течении которого интегрируется система
tstart, tend = 1, 100
time = [tstart, tend]
# Начальное условие
u0 = [-0.1, 5]
sol = solve_ivp(vanDerPol, time, u0, rtol = 1e-11, atol = 1e-11, dense_output=True)
# rtol, atol это абсолютная и относительная точность
# о том что это такое можно ознакомиться перейдя по ссылке, что указана выше
# так же и о том, что такое dense_output
```

```python
# message выводит сообщение о том успешно прошло ли интегрирование или же
# возникли какие-то неприятности, к примеру была обнаружена неустойчивость
# неустойчивость означает, что решение неустойчиво
# success указывает прошло ли успешно интегрирование, тут либо True либо False
# status здесь указывается цифра, которая соответствует в некотором роде
# сообщению, что вывелось для пользователя
# t это массив времени
# y массив точек
# все остальное в курсе не пригодится
sol
```

```python
ts = sol.t
xs, vs= sol.y
```
